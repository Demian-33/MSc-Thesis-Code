```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import normal_inverse_gamma
```

```python
# Hyperparameters
mu0 = 0       # prior mean
lambda0 = 1   # precision scaling for mu
a0 = 3        # shape for inverse-gamma
b0 = 2        # scale for inverse-gamma
```

```python
# Create a grid of mu and sigma^2 values
mu_vals = np.linspace(-5, 5, 300)
sigma2_vals = np.linspace(0.001, 1, 300)
MU, SIGMA2 = np.meshgrid(mu_vals, sigma2_vals)

# Evaluate joint density
rv = normal_inverse_gamma(mu=mu0, lmbda=lambda0, a=a0, b=b0)
Z = rv.pdf(MU, SIGMA2)
```

```python
# Plotting
plt.figure(figsize=(8, 8))
contour = plt.contourf(MU, SIGMA2, Z, levels=50, cmap='viridis')
plt.colorbar(contour, label='Density')
plt.title('Normal-Inverse-Gamma Density Contour')
plt.xlabel(r'$\mu$')
plt.ylabel(r'$\sigma^2$')
#plt.yscale('log')  # optional, shows detail in small sigma^2 values
plt.grid(True, which='both', ls='--', lw=0.5)
plt.tight_layout()
plt.show()
```

```python
# Plot level curves (contour lines)
plt.figure(figsize=(8, 8))
contour_lines = plt.contour(MU, SIGMA2, Z, levels=15, cmap='plasma')
plt.clabel(contour_lines, inline=True, fontsize=8)
plt.xlabel(r'$\mu$')
plt.ylabel(r'$\sigma^2$')
plt.title('Level Curves of Normal-Inverse-Gamma Density')
#plt.yscale('log')  # Helps visualize small sigma^2
plt.grid(True, which='both', ls='--', lw=0.5)
plt.tight_layout()
plt.show()
```

```python
# Set levels, excluding near-zero
Zmax = Z.max()
min_level = Zmax * 1e-3  # threshold: 0.1% of max
levels = np.linspace(min_level, Zmax, 15)

# Plot
plt.figure(figsize=(8, 6))
contours = plt.contour(MU, SIGMA2, Z, levels=levels, cmap='plasma')
plt.clabel(contours, inline=True, fmt="%.2f", fontsize=8)
plt.xlabel(r'$\mu$')
plt.ylabel(r'$\sigma^2$')
plt.title('Level Curves of Normal-Inverse-Gamma Density (No near-zero)')
#plt.yscale('log')  # Optional
plt.grid(True, which='both', ls='--', lw=0.5)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invgamma, norm
```

```python
# Sample data
np.random.seed(0)
n = 20
mu_true = 2
sigma2_true = 1.5
y = np.random.normal(mu_true, np.sqrt(sigma2_true), size=n)
ybar = np.mean(y)
sum_y2 = np.sum(y**2)

# Prior hyperparameters
mu0 = 0
lambda0 = 1
a0 = 3
b0 = 2

# Variational parameters
a_star = a0 + (n + 1) / 2

# Initial guess for b_star
mu_star = (n * ybar + lambda0 * mu0) / (n + lambda0)
mu2_star = mu_star**2
b_star = 1.0  # init

# Iterate to convergence (1-2 iterations usually enough)
for _ in range(5):
    sigma2_star = b_star / (a_star - 1) / (n + lambda0)
    E_mu2 = sigma2_star + mu2_star

    b_star = b0 + 0.5 * (
        (n + lambda0) * E_mu2
        - 2 * (n * ybar + lambda0 * mu0) * mu_star
        + sum_y2 + lambda0 * mu0**2
    )

# Final values
print(f"a*: {a_star:.3f}")
print(f"b*: {b_star:.3f}")
print(f"mu*: {mu_star:.3f}")
print(f"sigma^2*: {sigma2_star:.3f}")

# Plot variational approximation q(mu) * q(sigma^2)
mu_vals = np.linspace(mu_star - 4*np.sqrt(sigma2_star), mu_star + 4*np.sqrt(sigma2_star), 300)
sigma2_vals = np.linspace(0.01, 5, 300)
MU, SIGMA2 = np.meshgrid(mu_vals, sigma2_vals)

qmu = norm.pdf(MU, loc=mu_star, scale=np.sqrt(sigma2_star))
qs2 = invgamma.pdf(SIGMA2, a=a_star, scale=b_star)
Z = qmu * qs2
```

```python
# Contour plot
plt.figure(figsize=(8, 6))
plt.contour(MU, SIGMA2, Z, levels=15, cmap="cividis")
plt.xlabel(r'$\mu$')
plt.ylabel(r'$\sigma^2$')
plt.title('Variational Approximation: $q(\mu)q(\sigma^2)$')
#plt.yscale("log")
plt.grid(True, which="both", ls="--")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import invgamma, norm, normal_inverse_gamma
```

```python
# Set random seed
np.random.seed(123)

# Prior hyperparameters
mu0 = 0
lambda0 = 1
a0 = 3
b0 = 2

# Sample from hierarchical prior
sigma2_true = invgamma.rvs(a0, scale=b0)
mu_true = np.random.normal(mu0, np.sqrt(sigma2_true / lambda0))

# Generate data from likelihood
n = 200
y = np.random.normal(mu_true, np.sqrt(sigma2_true), size=n)
ybar = np.mean(y)
sum_y2 = np.sum(y**2)

# Grid for plotting
mu_vals = np.linspace(-1.1, -0.8, 300)
sigma2_vals = np.linspace(0.9, 1.3, 300)
MU, SIGMA2 = np.meshgrid(mu_vals, sigma2_vals)

# True posterior (joint)
#true_post = normal_inverse_gamma(mu=mu0, lmbda=lambda0, a=a0, b=b0)
#Z_true = true_post.pdf(MU, SIGMA2)

# Update true posterior hyperparameters
lambda_n = lambda0 + n
mu_n = (lambda0 * mu0 + n * ybar) / lambda_n
a_n = a0 + n / 2
sq_diff = np.sum((y - ybar)**2)
b_n = b0 + 0.5 * sq_diff + (lambda0 * n * (ybar - mu0)**2) / (2 * lambda_n)

# True posterior from data
true_post = normal_inverse_gamma(mu=mu_n, lmbda=lambda_n, a=a_n, b=b_n)
Z_true = true_post.pdf(MU, SIGMA2)

# Initialize variational parameters
mu_star = -0.9
a_star = 1.0
b_star = 2.0
sigma2_star = 0.25


# To save snapshots of variational approx
Z_variational_list = []

# Run variational updates and save approximation at various stages
for it in range(201):
    # Compute variational density every 5 iterations (including 0)
    if it in [0, 1, 20, 200]:
        qmu = norm.pdf(MU, loc=mu_star, scale=np.sqrt(sigma2_star))
        qs2 = invgamma.pdf(SIGMA2, a=a_star, scale=b_star)
        Z_q = qmu * qs2
        Z_variational_list.append(Z_q)
    
    # Update sigma2_star
    sigma2_star = b_star / (a_star * (n + lambda0)) 
    E_mu = mu_star
    E_mu2 = sigma2_star + mu_star**2

    # Update b_star
    b_star = b0 + 0.5 * (
        (n + lambda0) * E_mu2
        - 2 * (n * ybar + lambda0 * mu0) * E_mu
        + sum_y2 + lambda0 * mu0**2
    )
    
    mu_star = (n * ybar + lambda0 * mu0) / (n + lambda0) # fixed
    a_star = a0 + (n + 1) / 2 # fixed

```

```python
# Plot comparisons
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
titles = ['Iteration 0', 'Iteration 1', 'Iteration 20', 'Iteration 200']

for i, ax in enumerate(axes.flat):
    cs1 = ax.contour(MU, SIGMA2, Z_true, levels=5, cmap='Blues')
    cs2 = ax.contour(MU, SIGMA2, Z_variational_list[i], levels=5, cmap='Oranges')
    #ax.set_title(f"True Posterior vs Variational ({titles[i]})")
    # Add dummy handles for legend
    blue_line = plt.Line2D([0], [0], color='blue', linestyle='solid', label='Posterior exacta')
    red_line = plt.Line2D([0], [0], color='red', linestyle='dashed', label='Approx. IBV')
    ax.set_xlabel(r'$\mu$')
    ax.set_ylabel(r'$\sigma^2$')
    #ax.set_yscale('log')
    ax.legend(handles=[blue_line, red_line], loc='upper right', fontsize=10, frameon=True)
    ax.grid(True, which='both', ls='--', lw=0.5)


# Put the legend in a single spot
#fig.legend(handles=[blue_line, red_line], loc='lower center', ncol=2, fontsize=12)
#fig.suptitle("Normal-Inverse-Gamma Posterior vs Variational Approximation", fontsize=16)
plt.tight_layout(rect=[0, 0.03, 1, 0.97])
plt.savefig("normal-VB-contour.pdf", bbox_inches='tight')
plt.show()
```
